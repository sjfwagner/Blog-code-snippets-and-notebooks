```python
import pandas as pd
import dateutil.parser as parse
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
```

```python
#with open("Gender SYMIH - Sheet1.csv") as infile
```

```python
df = pd.DataFrame()
```

```python
symih = df.from_csv("UPDATE.csv",index_col=False)
#symih.Date = pd.to_datetime(symih.Date)
```

```python
symih.head()
```

```python
def check_fem(x):
    if x == 'F':
        return 1
    else: 
        return 0
def check_male(x):
    if x == 'M':
        return 1
    else: 
        return 0 
def check_neutral(x):
    if x == 'N':
        return 1
    else: 
        return 0
def check_both(x):
    if x == 'B':
        return 1
    else: 
        return 0
symih["FemaleLoose"] = map(check_fem,symih['Gender-Loose'])
symih["MaleLoose"] = map(check_male,symih['Gender-Loose'])
symih["BothLoose"] = map(check_both,symih['Gender-Loose'])
symih["NeutralLoose"] = map(check_neutral,symih['Gender-Loose'])

symih["MaleTight"] = map(check_male,symih['Gender-Tight'])
symih["FemaleTight"] = map(check_fem,symih['Gender-Tight'])
symih["NeutralTight"] = map(check_neutral,symih['Gender-Tight'])
symih["BothTight"] = map(check_both,symih['Gender-Tight'])
```

```python
def month_year(x):
    return "-".join([x.split("/")[0],'1',x.split("/")[2]])
symih['month_year'] = map(month_year,symih.Date)
```

```python
symih
```

```python
#symih.groupby(symih,by=symih.month_year)#.sum()
gender = symih.groupby('month_year')
#list(symih.Date)
```

```python
gender.describe()
```

```python
gender_line = gender.sum()
```

```python
#gender_line['LooseTotal']
```

```python
gender_line.index = map(parse.parse,gender_line.index)
```

```python
gender_line.sort(inplace=True)
```

```python
gender_line.to_clipboard()
```

```python
x = gender_line.index
y1 = gender_line.MaleLoose
y2 = gender_line.FemaleLoose
#y3 = [sum([i,j]) for i,j in y1,y2]
y3=[]
for i in range(len(y1)):
    y3.append(sum([y1[i],y2[i]])/2.0)
```

```python
y3
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(1)
fig.autofmt_xdate()
#ax.set_xbound(upper=2000)
#ax.set_xlim(right=2000)
#ax.set_xlim(left=1200)

x = gender_line.index
y1 = gender_line.MaleLoose
y2 = gender_line.FemaleLoose
y3=[]
for i in range(len(y1)):
    y3.append(sum([y1[i],y2[i]])/2.0)
    
fem_avg = float(sum(y2))/len(y2)
female_avg = []
for i in y2:
    female_avg.append(fem_avg)

ma_avg = float(sum(y1))/len(y1)
male_avg = []
for i in y1:
    male_avg.append(ma_avg)   
    
    
ax.plot(x,y1,linewidth=1, label = "Male Episodes", color = "blue")#,alpha=.75)

fig.set_size_inches(15, 8)
fig.set_dpi(1200)
#fig.set_tight_layout()
ax.plot(x,y2, linewidth = 1,label = "Female Episodes", color ="pink")#,alpha=.75)
ax.plot(x,male_avg,linewidth =4,label ="Avg Male Episodes",color = 'blue',linestyle=':')
ax.plot(x,female_avg,linewidth =4,label ="Avg Female Episodes",color = 'red',linestyle=':')

#ax.plot( x,y1,linewidth=1, label = "All Biographies")


        
ax.plot(x,y3,'-',linewidth=1,color="purple", label = "50% Mark",alpha=.75)

ax.fill_between(x,0,y2,alpha=.25, color="red")
ax.fill_between(x,y2,y1,alpha=.25)
ax.set_title("Number of 'Loose' Episodes by Month/Year")
ax.legend(loc = "upper left")
ax.set_xlabel("Month of Release")
ax.set_ylabel("Number of Episodes")


fname = "episodes by gender-month.jpg"
fig.savefig(fname, dpi=400, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None)

#fig.title("Number of Biographies by Birth Year")
```

```python
loose_male_count = gender_line.MaleLoose.sum()
loose_female_count = gender_line.FemaleLoose.sum()
loose_both_count = gender_line.BothLoose.sum()
loose_neutral_count = gender_line.NeutralLoose.sum()
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(1)

fig.set_size_inches(13, 13)
fig.set_dpi(1200)


ax.pie([loose_male_count,loose_female_count,loose_both_count,loose_neutral_count],
        labels = ['Male','Female','Both','Neutral'],
        explode=[0.032,0.032,0.032,0.032],
        colors=['lightblue','palevioletred','purple','lightgrey'],
        labeldistance=1.1)
ax.set_title("Number of 'Loose' Episodes by Gender")

fname = "loosepiechart.jpg"
fig.savefig(fname, dpi=400, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None)
```
