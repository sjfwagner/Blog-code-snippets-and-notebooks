```python
from lxml import html
import requests
```

```python
page = requests.get("http://www.missedinhistory.com/podcasts/archive/")
```

```python
archive = html.fromstring(page.content)
```

```python
values = []
for i in archive.xpath('//a[@class="title"]'):
    value = {
    'link': i.xpath('@href')[0],
    'title': i.text}
    values.append(value)
```

```python
keys = [j.text for j in archive.xpath('//p[@class="post_date_only"]')]
```

```python
data = dict(zip(keys,values))
```

```python
#<meta property="og:description" content="Pizza-like foods go way, way back in history, long before we associated the delicious dish with Italy. How did pizza's pedigree develop, and how did it get to its second home in the U.S.?" />
```

```python
page = requests.get("http://www.missedinhistory.com/podcasts/a-history-of-pizza-live-at-c2e2/")
page_description = html.fromstring(page.content)
```

```python
page_description.xpath('//meta[@property="og:description"]/@content')[0]
```

```python
for i in data: 
    page = requests.get(data[i]["link"])
    page_description = html.fromstring(page.content)
    data[i]['description'] = page_description.xpath('//meta[@property="og:description"]/@content')[0]
```

```python
data
```

```python
import pandas as pd
```

```python
x = pd.DataFrame.from_dict(data,orient='index')
```

```python
x
```

```python
x.link = map(lambda i: unicode(i),x.link)
x.description = map(lambda i: unicode(i),x.description)
x.title = map(lambda i: unicode(i),x.title)
```

```python
x.index = map(lambda i: unicode(i), x.index)
```

```python
x.to_clipboard(encoding='utf-8')
```
