```python
def get_dissertation_content(diss):
    dissertator = diss.split('<b>')[1].split(':</b>')[1].split("</")[0].strip()
    title = diss.split('<b>')[3].split('</b>')[1].split('<br>')[0].strip()
    if 'None given' in diss.split('<b>')[3].split('</B>')[1]:
        advisor = 'None given'
    else: 
        advisor = diss.split('<b>')[3].split('</B>')[1].split("<br>")[0].strip()
    completed = diss.split('<b>')[3].split("Completed:")[1].split("<br>")[0].strip()
    return [dissertator,title,advisor,completed]
```
